```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import os
```

```python
met = pd.read_json('data/ground_truth/MET_database.json')
mini_met = pd.read_json('data/ground_truth/mini_MET_database.json')
truth_test = pd.read_json('data/ground_truth/testset.json')
truth_val = pd.read_json('data/ground_truth/valset.json')
```

```python
met
```

```python
ids = met["id"]
count = defaultdict(int)
for img in ids:
    count[img] += 1
```

```python
count_df = pd.DataFrame.from_dict(count, orient = 'index')
count_df.columns = ['count']
count_df.sort_values('count', ascending = False, inplace = True)
```

```python
count_df.describe()
```

```python
# fig, ax = plt.subplots()
# ax = sns.barplot(x = count_df.index, y = count_df['count'], ax = ax)
sns.countplot(x = 'id',
              data = met,
              order = met['id'].value_counts().iloc[:50000].index)
plt.show()
```

```python
from os import listdir
from PIL import Image as PImage

def loadImages(path):
    imagesList = listdir(path)
    loadedImages = []
    for image in imagesList:
        img = PImage.open(path + image)
        loadedImages.append(img)
    return loadedImages

path = "data/test_met/"

imgs = loadImages(path)

fig = plt.gcf()
for img in imgs[:10]:
    plt.imshow(img)
    plt.show()
```
